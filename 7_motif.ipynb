findMotifsGenome.pl  
HOMERツールを使用して、MACS2で検出されたピーク領域の中心付近（summits）でDNAモチーフの発見とエンリッチメント解析を行う。  
入力ファイル、使用するリファレンスゲノム、出力先のディレクトリはそのまま記載する。  
--size 200 各ピークの中心から上下100bpを対象にする。デフォルトは領域全体が解析される。   
-mask　繰り返し配列を無視して解析する。  
-p 12 モチーフ解析を12スレッドで実行する。

```python
! findMotifsGenome.pl \
    /home/inou_doi/demo/3_peak_call/MACS2/SRR7475870_summits.bed \
    hg19 \
    SRR7475870_motif_macs2  \
    -size 200 \
    -mask \
    -p 12
```
